```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
df = pd.read_csv("K:/DATA SCIENCE/DataSets/imdb_movies.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
print("Number of movies is: ", df.shape[0]) 
```

```python
df_described = df.describe().T.style.format('{:,.2f}')
display(df_described)
```

```python
corr = df[["score", "revenue", "budget_x"]].corr()  # Correlation between "score", "revenue", "budget_x"
sns.heatmap(corr, annot=True)
```

```python
top_10_revenues = df.sort_values("revenue", ascending=False)[:10]
top_10_revenues
```

```python
plt.bar(top_10_revenues['names'],(top_10_revenues['revenue']/100000000))
plt.title('Revenue of Top 10 Movies')
plt.xlabel('Movie_Name')
plt.ylabel('Revenue in Hundred Million')
plt.xticks(rotation=90)
plt.show()
```

```python
revenue_by_country = df.loc[:, ['country', 'revenue']]
revenue_by_country = revenue_by_country.groupby('country').sum()
top_5_countries = revenue_by_country.sort_values("revenue", ascending=False)[:5]

plt.bar(top_5_countries.index, top_5_countries['revenue']/1000000000)

plt.xticks(rotation=90)
plt.yticks(rotation=90)

plt.xlabel('Country')
plt.ylabel('Revenue (in billions)')

plt.show()
```

```python
df['date_x'] = pd.to_datetime(df['date_x'])
df['year'] = df['date_x'].dt.year

mean_by_year = df.groupby('year')[['revenue', 'budget_x']].mean()
```

```python
plt.figure(figsize=(30,15))
plt.bar(mean_by_year.index, mean_by_year.revenue/1000000, width=1)
plt.xlabel("Year", fontsize=25)
plt.ylabel("Revenue (in millions)", fontsize=25)
plt.title("Revenue over Years (1903-2023)", fontsize=25)
```

```python
revenue_by_language = df.loc[:, ['orig_lang', 'revenue']]
revenue_by_language = revenue_by_language.groupby('orig_lang').sum()
top5_revenuesbylanguage = revenue_by_language.nlargest(5,'revenue')
print(top5_revenuesbylanguage)
```
